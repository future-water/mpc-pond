```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 8400
t_range = np.linspace(0, 8400, N)
u = 0.15 * np.sin(t_range / 3600)

p2 = 0.23 * np.sin( (t_range ) / 3600 ) + 0.15 * np.sin(1200 / 3600)
p23 = -0.05 * np.sin( (t_range ) / 1200 ) + 0.15 * np.sin(1200 / 3600)
p24 = 0.28 * np.sin( (t_range ) / 2900 ) + 0.15 * np.sin(1200 / 3600)

p3 = -0.025 * np.sin( (t_range ) / 1200 ) + 0.23 * np.sin( (1200 ) / 3600 ) + 0.15 * np.sin(1200 / 3600)
p32 = 0.22 * np.sin( (t_range ) / 1900 ) + 0.23 * np.sin( (1200 ) / 3600 ) + 0.15 * np.sin(1200 / 3600)
p33 = 0.26 * np.sin( (t_range ) / 2800 ) + 0.23 * np.sin( (1200 ) / 3600 ) + 0.15 * np.sin(1200 / 3600)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.set_palette('husl', 4)
plt.axvspan(0, 4800, facecolor='dimgray', alpha=0.2)
plt.axvspan(1200, 6000, facecolor='darkgray', alpha=0.2)
plt.axvspan(2400, 7200, facecolor='lightgray', alpha=0.2)
plt.axvspan(3600, 8400, facecolor='whitesmoke', alpha=0.2)

plt.plot(t_range[1:1200] + 1200, p2[1:1200], c ='dodgerblue', linewidth = 2.5)
plt.plot(t_range[1:4800] + 1200, p2[1:4800], '--', c ='grey',alpha=0.7)
plt.plot(t_range[1:1200] + 1200, p23[1:1200], c = 'grey')
plt.plot(t_range[1:4800] + 1200, p23[1:4800], '--', c ='grey',alpha=0.7)
plt.plot(t_range[1:1200] + 1200, p24[1:1200], c = 'grey')
plt.plot(t_range[1:4800] + 1200, p24[1:4800], '--', c ='grey',alpha=0.7)

plt.plot(t_range[1:1200] + 2400, p3[1:1200],c ='slategrey')
plt.plot(t_range[1:4800] + 2400, p3[1:4800], '--', c ='slategrey',alpha=0.7)
plt.plot(t_range[1:1200] + 2400, p32[1:1200],c ='slategrey')
plt.plot(t_range[1:4800] + 2400, p32[1:4800], '--', c ='slategrey',alpha=0.7)
plt.plot(t_range[1:1200] + 2400, p33[1:1200],c ='dodgerblue', linewidth = 2.5)
plt.plot(t_range[1:4800] + 2400, p33[1:4800], '--', c ='slategrey',alpha=0.7, label = 'Suboptimal paths')

plt.plot(t_range[1:4800], u[1:4800], '--', c ='grey', alpha=0.7)
plt.plot(t_range[1:1200], u[1:1200], c ='dodgerblue', linewidth = 2.5, label = 'Optimal path')

plt.xlim([-600, 8400])
plt.xticks([0, 1200, 2400, 3600, 4800, 6000, 7200], [r'$ \tau $',r'$ \tau +1 $',r'$ \tau+2 $','',r'$ \tau+T $',r'$ \tau+T+1 $',r'$ \tau+T+2 $'])
plt.yticks([])
plt.ylim([-0.05,.4])
ax.margins(x=0, y=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.ylabel('States')
plt.legend(frameon=False, loc = 'lower right')
plt.savefig('mpc_receding_horizon.png', dpi=400)
plt.show()
```
