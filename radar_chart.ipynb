```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
subjects=['Overflow','Peak outflow','TSS Load','Control effort', 'Outflow flashiness']
MPC=[0,52.0813,13.7164,12.5027,57.1292]
RBC_c=[0,100,10.1137,100,100]
RBC_q = [0,52.0813,35.0775,15.3913,63.0609]
Passive = [0, 73.4538, 100, 0, 74.4027]
negative = [-10, -10, -10, -10, -10]
```

```python
angles=np.linspace(0,2*np.pi,len(subjects), endpoint=False)
print(angles)
```

```python
angles=np.concatenate((angles,[angles[0]]))
subjects.append(subjects[0])
MPC.append(MPC[0])
RBC_c.append(RBC_c[0])
RBC_q.append(RBC_q[0])
Passive.append(Passive[0])
negative.append(negative[0])
```

```python
subjects2=['Overflow','Peak outflow','TSS Load','Control effort', 'Outflow flashiness']
MPC2=[0,51.3163,1.4583,4.1850,30.9491]
RBC_c2=[36.6562,100,20.2039,100,100]
RBC_q2 = [100,51.3163,100,5.1077,54.5951]
Passive2 = [0, 72.4242, 8.3778, 0, 60.8208]
negative2 = [-10, -10, -10, -10, -10]
```

```python
angles2=np.linspace(0,2*np.pi,len(subjects2), endpoint=False)
```

```python
angles2=np.concatenate((angles2,[angles2[0]]))
subjects2.append(subjects2[0])
MPC2.append(MPC2[0])
RBC_c2.append(RBC_c2[0])
RBC_q2.append(RBC_q2[0])
Passive2.append(Passive2[0])
negative2.append(negative2[0])
```

```python
fig=plt.figure(figsize=(12,6))
ax = fig.add_subplot(121, projection='polar')
ax2 = fig.add_subplot(122, polar=True)

#MPC plot
ax.plot(angles,Passive, 'o-', color='k', label='Passive')
ax.fill(angles, Passive, alpha=0.25, color='k')

ax.plot(angles,MPC, 'o-', color='royalblue', label='MPC')
ax.fill(angles, MPC, alpha=0.25, color='royalblue')

#RBC_c Plot
ax.plot(angles,RBC_c, 'o-', color='orange', linewidth=1, label='RBC-Conc.')
ax.fill(angles,RBC_c, alpha=0.25, color='orange')

#RBC_q Plot
ax.plot(angles,RBC_q, 'o-', color='orangered', linewidth=1, label='RBC-Outflow')
ax.fill(angles,RBC_q, alpha=0.25, color='orangered')

ax.plot(angles,negative, 'o-', alpha=0, color='orangered', linewidth=1, label='')
ax.fill(angles,negative, alpha=0, color='orangered')
ax.set_thetagrids(angles * 180/np.pi, subjects)
##2
#MPC plot
ax2.plot(angles2,Passive2, 'o-', color='k', label='Passive')
ax2.fill(angles2, Passive2, alpha=0.25, color='k')

ax2.plot(angles2,MPC2, 'o-', color='royalblue', label='MPC')
ax2.fill(angles2, MPC2, alpha=0.25, color='royalblue')

#RBC_c Plot
ax2.plot(angles2,RBC_c2, 'o-', color='orange', linewidth=1, label='RBC-Conc.')
ax2.fill(angles2,RBC_c2, alpha=0.25, color='orange')

#RBC_q Plot
ax2.plot(angles2,RBC_q2, 'o-', color='orangered', linewidth=1, label='RBC-Outflow')
ax2.fill(angles2,RBC_q2, alpha=0.25, color='orangered')

ax2.plot(angles2,negative2, 'o-', alpha=0, color='orangered', linewidth=1, label='')
ax2.fill(angles2,negative2, alpha=0, color='orangered')

#Add labels
ax2.set_thetagrids(angles2 * 180/np.pi, subjects)
ax.set_title('(a)', loc='left')
ax2.set_title('(b)', loc='left')

plt.grid(True)
plt.tight_layout()
plt.plot()
plt.legend(frameon=False, loc = 'lower right')
plt.savefig('performance_pond4.png', dpi=300)
plt.show()
```
